```python
from dsa.constants import ROOT
from dsa.utils import load_json
import pandas as pd
```

```python
res_dir = ROOT / "data/evaluation_output/"
# res_files = sorted(list(res_dir.glob("report_*.json")))
fnames = [
    "report_tfid-large_unhcr_combined_wo_small.json",
    "report_tfid-large_prwp_batch1_wo_small.json",
    "report_tfid-large_refugee_batch1_wo_small.json",
    "report_doclayoutyolo_unhcr_combined_wo_small.json",
    # "report_doclayoutyolo_prwp_batch1_wo_small.json",
    "report_doclayoutyolo_prwp_batch1.json",
    # "report_doclayoutyolo_refugee_batch1_wo_small.json",
    "report_doclayoutyolo_refugee_batch1.json",
    "report_yolo11_unhcr_combined_wo_small.json",
    # "report_yolo11_prwp_batch1_wo_small.json",
    "report_yolo11_prwp_batch1.json",
    # "report_yolo11_refugee_batch1_wo_small.json",
    "report_yolo11_refugee_batch1.json",
    "report_yolo26_unhcr_combined_wo_small.json",
    "report_yolo26_prwp_batch1_wo_small.json",
    "report_yolo26_refugee_batch1_wo_small.json",
]
res_files = [res_dir / x for x in fnames]

res_files
```

```python
for f in res_files:
    f.resolve(strict=True)
```

```python
thresholds = [
    "0.5",
    # "0.75",
]
classes = [
    "Figure",
    "Table",
]
metrics = [
    "precision",
    "recall",
    "mean_iou",
    "mean_area_precision",
    "mean_area_recall",
]
```

```python
cols = []
for r in res_files:
    row = {}
    row["filename"] = r.name

    res = load_json(r)
    for t in thresholds:
        for c in classes:
            for m in metrics:
                row[f"[{c}] {m} @ {t}"] = res["metrics"][t]["per_class"][c][m]

    cols.append(row)
```

```python
df = pd.DataFrame(cols).set_index("filename").transpose()
df
```
