```python
import glob
import pandas as pd
```

# Load files

```python
files = glob.glob("data/results_*.jsonl")
files
```

```python
df = pd.concat([pd.read_json(x, lines=True) for x in files]).reset_index(drop=True)
df
```

```python
# df = df[df["model"] == "gpt-5.4-mini"].copy()
```

# Type

```python
df["type"] = df["snapshot_id"].apply(lambda x: x.split("_")[-2])
df.groupby(["source", "type"]).agg(
    count=("source", "count"),
)
```

# Cost

```python
df = df.reset_index(drop=True).join(pd.json_normalize(df["cost"]))

total = df["total_cost_usd"].sum()
mean = df["total_cost_usd"].mean()
std = df["total_cost_usd"].std()

print(f"Total cost: ${total:.2f}")
print(f"Ave cost per snapshot: ${mean:.4f}")
print(f"Std dev: ${std:.4f}")
```

```python
df.groupby(["source", "model"]).agg(
    count=("source", "count"),
    total_cost_usd=("total_cost_usd", "sum"),
    ave_cost_usd=("total_cost_usd", "mean"),
    std_cost_usd=("total_cost_usd", "std"),
)
```

# Output tokens

```python
total = df["output_tokens"].sum()
mean = df["output_tokens"].mean()
std = df["output_tokens"].std()

print(f"Total output tokens: {total}")
print(f"Ave output token per snapshot: {mean}")
print(f"Std dev: {std:.2f}")
```

```python
df.groupby(["source", "model"]).agg(
    count=("source", "count"),
    total_output_tokens=("output_tokens", "sum"),
    ave_output_tokens=("output_tokens", "mean"),
    std_output_tokens=("output_tokens", "std"),
)
```
